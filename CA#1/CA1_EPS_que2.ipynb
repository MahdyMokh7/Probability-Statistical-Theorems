1. import the needed libraries


```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson, norm, binom


N = 250
P = 0.008
SIZE = 1000
```

2. draw_function
  gets the type that is either cdf or pdf and it will draw binomial, poisson,and normal distribution of the respective type

  we can compare the distributions and their approximation

```python
def draw_chart(bin_samples, type_name):
    # create figure
    plt.figure()

    # create x-axis range
    x_value_range = np.arange(0, 6, 1)
    if type_name == 'cdf':  # generate plot cdf
        plot_cdf = plt.plot(x_value_range, binom.cdf(x_value_range, N, P)
                            , color='b', linewidth=1, label='Binomial')
        plot_cdf = plt.plot(x_value_range, poisson.cdf(x_value_range, N*P)
                                , color='r', linewidth=1, label='Poisson')
        plot_cdf = plt.plot(x_value_range, norm.cdf(x_value_range, N*P, math.pow(N*P*(1-P), 1/2))
                               , color='g', linewidth=1, label='Normal')

    if type_name == 'pdf':  # generate plot pdf/pmf
        plot_pdf = plt.plot(x_value_range, binom.pmf(x_value_range, N, P)
                            , color='b', linewidth=1, label='Binomial')
        plot_pdf = plt.plot(x_value_range, poisson.pmf(x_value_range, N*P)
                                , color='r', linewidth=1, label='Poisson')
        plot_pdf = plt.plot(x_value_range, norm.pdf(x_value_range, N*P, math.pow(N*P*(1-P), 1/2))
                               , color='g', linewidth=1, label='Normal')

    # configure
    plt.xlabel('Samples')
    plt.ylabel(type_name)
    plt.title('Comparison')
    plt.legend()

    # draw graph
    plt.show()
```

3. calling the funciton for each pdf and cdf

```python
draw_chart(np.random.binomial(n=250, p=0.008, size=SIZE), type_name='cdf')
draw_chart(np.random.binomial(n=250, p=0.008, size=SIZE), type_name='pdf')
```

As you can see from the graphs above, the Poisson distribution is a much better approximation of the Binomial distribution than the Normal distribution

A. becuase n -> $\infty$ and p isn't equal to 1/2 or even near 1/2; and in this case Normal dist. isn't a good approximation of the Binomial dist.
B. becuase n -> $\infty$ and p -> 0 the Poisson dist. is a great approximation of the binomial dist.

C. The reason that we use these approximations in the first placce instead of just using the binmomial is because sometimes it gets hard to calculate the wanted probability with the binomial dist. so we convert it to either Normal or Poisson dist.

d. How to find out which is the better option?

  1. n -> $\infty$ & $p \approx 1/2$: Normal dist.

  2. n -> $\infty$ & p -> 0: Poisson dist.
